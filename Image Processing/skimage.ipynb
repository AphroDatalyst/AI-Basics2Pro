```python
import matplotlib.pyplot as plt
img_beach = plt.imread(".\sample1.JPG")
plt.imshow(img_beach)
plt.title("Beach")
plt.axis("off")
plt.show()
```

```python
image1 = data.cat()
image2 = data.coffee()
image3 = data.rocket()
image4 = data.checkerboard()
image5 = data.camera()
image6 = data.astronaut()

titles = ["cat","coffee","rocket","checkerboard","camera","astronaut"]
images = [image1,image2,image3,image4,image5,image6]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")
plt.show()
```
